```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("career_path_matrix.csv")
```

```python
df
```

```python
career_list = df['prerequisites'].tolist()
career_list
```

```python
user_skills = "python,flask,sql"
# user_skills = (",").join("selected_skill")
```

```python
document= career_list + [user_skills]
document
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(document)
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(
    tfidf_matrix[-1],
    tfidf_matrix[:-1]
)
```

```python
similarity_scores
```

```python
best_match = similarity_scores.argmax()
best_match
```

```python
career_list[best_match]
```

```python
match =similarity_scores.argsort()
```

```python
match
```

```python
ranked_indices = similarity_scores.argsort()[::-1].flatten()
```

```python
top_indices = ranked_indices[:5]
top_indices
```

```python
for i in range(5):
    print(career_list[top_indices[i]])
```

```python
df.iloc[top_indices]
```

```python
for index in top_indices:
    career = df.iloc[index]

    # print("Career:", career["career"])
    print("Skills:", career["career_path"])
    print("Prerequisites:", career["prerequisites"])
    print("-----------------------")
```

```python
similarity_scores = similarity_scores.flatten()
```

```python
top_indices
```

```python
print(similarity_scores)
```

```python
# Create user skill set once
user_skill_set = {skill.strip().lower() for skill in user_skills.split(",")}

for index in top_indices:

    career = df.iloc[index]

    score = similarity_scores[index] * 100

    # Required skills for this career
    required_skill_set = set(career["prerequisites"].split(","))

    # Skill Gap Analysis
    matched_skills = user_skill_set & required_skill_set
    missing_skills = required_skill_set - user_skill_set

    matched_count = len(matched_skills)
    required_count = len(required_skill_set)

    # Display Recommendation
    print("=" * 60)
    print("Career:", career["career_path"])
    print("Match Score:", round(score, 2), "%")

    print(f"\nWhy this match?")
    print(f" Matched {matched_count} out of {required_count} required skills.")

    print("\nSkills You Already Have:")
    if matched_skills:
        print(", ".join(sorted(matched_skills)))
    else:
        print("None")

    print("\n Skills You Need to Learn:")
    if missing_skills:
        print(", ".join(sorted(missing_skills)))
    else:
        print("None! You already have all the required skills.")

    print("=" * 60)
    if score >= 80:
        print("\n⭐ Excellent Match!")
        print("Your profile strongly aligns with this career.")
        print("You already possess most of the required skills.")
        print("Continue learning the remaining skills to become job-ready.")

    elif score >= 60:
        print("\n👍 Good Match!")
        print("You have a solid foundation for this career.")
        print("Learning the missing skills will significantly improve your profile.")

    else:
        print("\n📚 Learning Path Recommended!")
        print("This career is achievable, but you'll need to build several core skills.")
        print("Focus on the missing skills listed below.")
```

```python
# Create user skill set once
user_skill_set = {skill.strip().lower() for skill in user_skills.split(",")}

# Store all recommendations
recommendations = []

for index in top_indices:

    career = df.iloc[index]
    score = similarity_scores[index] * 100

    # Required skills
    required_skill_set = set(career["prerequisites"].split(","))

    # Skill Gap Analysis
    matched_skills = user_skill_set & required_skill_set
    missing_skills = required_skill_set - user_skill_set

    matched_count = len(matched_skills)
    required_count = len(required_skill_set)

    # AI Recommendation Message
    if score >= 80:

        if missing_skills:
            message = (
                f"⭐ Excellent Match!\n"
                f"Your profile strongly aligns with this career.\n"
                f"You already have {matched_count} out of {required_count} required skills.\n"
                f"Focus on learning: {', '.join(sorted(missing_skills))}"
            )
        else:
            message = (
                f"⭐ Excellent Match!\n"
                f"Amazing! You already possess all the required skills for this career."
            )

    elif score >= 60:

        message = (
            f"👍 Good Match!\n"
            f"You already have {matched_count} out of {required_count} required skills.\n"
            f"Learning these skills will strengthen your profile:\n"
            f"{', '.join(sorted(missing_skills))}"
        )

    else:

        message = (
            f"📚 Learning Path Recommended!\n"
            f"This career is achievable, but you'll need to build several core skills.\n"
            f"You currently have {matched_count} out of {required_count} required skills.\n"
            f"Start learning:\n"
            f"{', '.join(sorted(missing_skills))}"
        )

    # Store recommendation
    recommendation = {
        "career": career["career_path"],
        "score": round(score, 2),
        "message": message,
        "matched_count": matched_count,
        "required_count": required_count,
        "matched_skills": sorted(matched_skills),
        "missing_skills": sorted(missing_skills),
    }

    recommendations.append(recommendation)
```
